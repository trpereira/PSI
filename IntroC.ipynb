## Introdução à Programação 


## Definições

### Algoritmia


#### Algoritmo

Conjunto de operações ordenadas e sequenciais que definem uma solução para um problema computacional.

#### Fluxograma

Representação gráfica de um algoritmo em que se verifica o fluxo do algorimto: apenas um início e apenas um fim. 


#### Pseudo-código

Conjunto de instruções que definem as operações de uma algoritmo em linguagem natural.

#### Código-Fonte

Conjunto de instruções que definem as operações de uma algoritmo em linguagem de programação.

#### Compilador

Software que interpreta o código fonte e converte em instruções binárias dando origem ao código objeto que pode ser executado num sistema operativo. Por esta razão, o compilador tem de ser específico para cada sistema operativo. Quando numa aplicação são utilizados vários códigos fonte, é necessário um ligador (linking) a seguir à compilação.

#### Código Objeto

Conjunto de instruções que definem as operações de uma algoritmo em linguagem binária.





# Linguagem de Programação C

## Estrutura de um programa

__Notação BNF__:

< Obrigatório, \[ Opcional, ::= Composto por, | Alternativa


### Programa ::= <Cabeçalho> [<função>] <_corpo>

### Tipo_de_dados ::= <_tipo> <_nome> [=valor inicial] <;>

### Cabeçalho ::= [<comentário>] [<Instruções de pré-processamento>] [Definição de funções] [<definição de variáveis globais>]


### Corpo ::= <tipo de dados de retorno> <main([<parâmetros>]){> <instrução;> <}>

### Função ::= <_tipo de dados de retorno ou void> <nome da função([<parâmetros>])\{> <instrução;> <\}>

### Parâmetros ::= <tipo_de_dados> <nome_>

### Instrução ::= <atribuição> | <decisão> | <repetição>

# Programa em C

```c
#include<stdio.h>
#include<string.h>

int max2i(int x, int y);
char *input_string="25,67";  //<-----input value

int main()
{
    int a,b,c;
    printf("input 2 integer numbers:");
    sscanf(input_string, "%d,%d", &a, &b);  //<------use sscanf instead of scanf
    c=max2i(a,b);
    printf("max value of (%d,%d) is %d\n", a,b,c);
    return 0;
}

int max2i(int x,int y)
{
    if(x>y)
        return x;
    else
        return y;
}
```

# Controlo de fluxo

## Condicionais

### if

CondiçãoIF ::= if(<condiçãodo ciclo>){<instruções>} [;] [else <instruções>];

```c
//If..else
#include <stdio.h>

int main(){
    int a=1, b=0;
    if(a==0){
        a=1;
        b=0;
    }
    else{
        a=0;
        b=1;
    }
    printf("valor de a:%d valor de b:%d", a, b );
 return(0);
}
```

### switch

switch(<expressão>){
    case <caso1>: <instrução><;>
    break;
    [case <caso2>;
    break; ...]
    default:<instrução>;
}
__NOTA:__ A expressão tem de ser sequencial 1,2,3,4,... ou A,B,C,D,... ou do tipo Enum

```c
//switch
#include <stdio.h>

 enum estado {aberto=1, fechado, obras};
 enum estado op;
int main(){
    op=aberto;
    int luz=0;
    switch(op){
        case aberto:
            luz=1;
        break;
        case fechado:
            luz=0;
        break;
        case obras:
            luz=0;
        break;
    default: luz=0;        
    }
 
    printf("A luz está (1- ligada/ 0 - desligada):%d", luz);
 return(0);
}
```
