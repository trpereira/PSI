## Introduo  Programao 


## Definies

### Algoritmia


#### Algoritmo

Conjunto de operaes ordenadas e sequenciais que definem uma soluo para um problema computacional.

#### Fluxograma

Representao grfica de um algoritmo em que se verifica o fluxo do algorimto: apenas um incio e apenas um fim. 


#### Pseudo-cdigo

Conjunto de instrues que definem as operaes de uma algoritmo em linguagem natural.

#### Cdigo-Fonte

Conjunto de instrues que definem as operaes de uma algoritmo em linguagem de programao.

#### Compilador

Software que interpreta o cdigo fonte e converte em instrues binrias dando origem ao cdigo objeto que pode ser executado num sistema operativo. Por esta razo, o compilador tem de ser especfico para cada sistema operativo. Quando numa aplicao so utilizados vrios cdigos fonte,  necessrio um ligador (linking) a seguir  compilao.

#### Cdigo Objeto

Conjunto de instrues que definem as operaes de uma algoritmo em linguagem binria.





# Linguagem de Programao C

## Estrutura de um programa

__Notao BNF__:

< Obrigatrio, \[ Opcional, ::= Composto por, | Alternativa


### Programa ::= <Cabealho> [<funo>] <_corpo>

### Tipo_de_dados ::= <_tipo> <_nome> [=valor inicial] <;>

### Cabealho ::= [<comentrio>] [<Instrues de pr-processamento>] [Definio de funes] [<definio de variveis globais>]


### Corpo ::= <tipo de dados de retorno> <main([<parmetros>]){> <instruo;> <}>

### Funo ::= <_tipo de dados de retorno ou void> <nome da funo([<parmetros>])\{> <instruo;> <\}>

### Parmetros ::= <tipo_de_dados> <nome_>

### Instruo ::= <atribuio> | <deciso> | <repetio>

# Programa em C

```c
#include<stdio.h>
#include<string.h>

int max2i(int x, int y);
char *input_string="25,67";  //<-----input value

int main()
{
    int a,b,c;
    printf("input 2 integer numbers:");
    sscanf(input_string, "%d,%d", &a, &b);  //<------use sscanf instead of scanf
    c=max2i(a,b);
    printf("max value of (%d,%d) is %d\n", a,b,c);
    return 0;
}

int max2i(int x,int y)
{
    if(x>y)
        return x;
    else
        return y;
}
```

## Condicionais

### if

CondioIF ::= if(<condiodo ciclo>){<instrues>} [;] [else <instrues>];

```c
//If..else
#include <stdio.h>

int main(){
    int a=1, b=0;
    if(a==0){
        a=1;
        b=0;
    }
    else{
        a=0;
        b=1;
    }
    printf("valor de a:%d valor de b:%d", a, b );
 return(0);
}
```

### switch

switch(<expresso>){
    case <caso1>: <instruo><;>
    break;
    [case <caso2>;
    break; ...]
    default:<instruo>;
}
__NOTA:__ A expresso tem de ser sequencial 1,2,3,4,... ou A,B,C,D,... ou do tipo Enum

```c
//switch
#include <stdio.h>

 enum estado {aberto=1, fechado, obras};
 enum estado op;
int main(){
    op=aberto;
    int luz=0;
    switch(op){
        case aberto:
            luz=1;
        break;
        case fechado:
            luz=0;
        break;
        case obras:
            luz=0;
        break;
    default: luz=0;        
    }
 
    printf("A luz est (1- ligada/ 0 - desligada):%d", luz);
 return(0);
}
```
